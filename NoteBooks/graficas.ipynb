```python
import numpy as np
import matplotlib.pyplot  as plt
import scienceplots

plt.style.use('science')
```

```python
def Ng(t,N_sat):  # En esta función simulamos la creación de trampas durante la irradiación
    C = 1
    return np.exp(np.log(N_sat/ 1) * (1 - np.exp(-t)))

plt.plot(Ng(t,10))
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.plot(t, Ng(t,10), label='$N_s$=10')
plt.plot(t, Ng(t,7), label='$N_s$=7')
plt.plot(t, Ng(t,5), label='$N_s$=5')
plt.plot(t, Ng(t,2), label='$N_s$=2')
plt.plot(t, Ng(t,1), label='$N_s$=1')
plt.title(r'$N_i$(t)')
plt.xlabel('t(s)')
plt.ylabel('$N_i$')
plt.legend()
plt.show()
```

```python
def N_i_fcrea(t,E,T):
    k = 0.00008617333262
    log=np.exp(np.log(10 / 1) * (1 - np.exp(-1 * t)))
    fd = (1+np.exp((E-1)/(k*T)))**-1
    return log*fd
t = np.linspace(0,10)

plt.figure(figsize=(7,5))
plt.plot(t, N_i_fcrea(t,1,300), label='E=1$E_f$')
plt.plot(t, N_i_fcrea(t,0.97,300), label='E=0.97$E_f$')
plt.plot(t, N_i_fcrea(t,0.93,300), label='E=0.93$E_f$')
plt.plot(t, N_i_fcrea(t,0.78,300), label='E=0.78$E_f$')
# plt.plot(t, N_i_fcrea(t,0.5,300), label='E=0.5$E_f$')
plt.title('$N_i(t,E)$')
plt.xlabel('t(s)')
plt.ylabel('$N_i$')
plt.legend()
plt.show()

plt.plot(t, N_i_fcrea(t,0.3,300), label='E=0.8$E_f$')
plt.show()
plt.plot(t, N_i_fcrea(t,0.5,300), label='E=0.8$E_f$')
plt.show()
plt.plot(t, N_i_fcrea(t,2,300), label='E=0.8$E_f$')
plt.show()
```
