```python
import numpy as np
import matplotlib.pyplot as plt

E_f = 1
E_c = 2
E_v = 0

k = 0.00008617333262

T = 300

# Valores de energía y probabilidad de ocupación
energy = np.linspace(-2, 2, 100)
fermi_dirac = 1 / (1 + np.exp((energy - E_f) / (k * T)))  # E_f y T son los valores deseados

# Valores de energía de la banda de conducción y de valencia
conduction_band = np.ones_like(energy) * E_c  # E_c es la energía de la banda de conducción
valence_band = np.ones_like(energy) * E_v  # E_v es la energía de la banda de valencia

# Configuración de la figura y los ejes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Gráfico de la banda de conducción y de valencia
ax1.plot([0, 1], [E_c, E_c], color='salmon', label='Banda de conducción')
ax1.plot(fermi_dirac, energy, color='green', label='Distribución de Fermi-Dirac')
ax1.plot([0, 1], [E_v, E_v], color='skyblue', label='Banda de valencia')
ax1.set_ylim(-0.1, 2.1)
ax1.set_xlim(0, 1)
ax1.set_xlabel('Sin defectos en la red')
ax1.set_ylabel('Energía (eV)')
ax1.legend()

# Gráfico de la distribución de Fermi-Dirac
ax2.plot(fermi_dirac, energy, color='green')
ax2.plot([0, 1], [E_c, E_c], color='salmon')
ax2.plot([0, 1], [1.5, 1.5], label='Trampa')
ax2.plot([0, 1], [0.5, 0.5], label='Centro de recombinación')
ax2.plot([0, 1], [E_v, E_v], color='skyblue')
ax2.set_ylim(-0.1, 2.1)
ax2.set_xlim(0, 1)
ax2.set_xlabel('Con defectos en la red')
ax2.set_ylabel('Energía (eV)')
ax2.legend()
# Mostrar la gráfica
plt.show()
```
